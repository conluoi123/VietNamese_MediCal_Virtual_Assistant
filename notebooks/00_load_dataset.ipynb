```python
from datasets import load_dataset
import os 
dataset = load_dataset("hungnm/vietnamese-medical-qa")

# Kiểm tra kết quả 
print(dataset['train'][0])
```

## Lưu kết quả vào thư mục để sử dụng

```python
DATA_PATH = "../data/raw"

if not os.path.exists(DATA_PATH):
    os.makedirs(DATA_PATH)
    print(f"Đã tạo thư mục Raw!")


try:
    if hasattr(dataset, 'keys'): 
        for split in dataset.keys():
            file_name = f"{split}.csv"
            save_to = os.path.join(DATA_PATH, file_name)
            dataset[split].to_csv(save_to, index=False)
            print(f"Đã lưu {split} tại: {save_to}")
    else: # Nếu là Dataset đơn lẻ
        save_to = os.path.join(DATA_PATH, "dataset.csv")
        dataset.to_csv(save_to, index=False)
        print(f"Đã lưu dataset tại: {save_to}")
except Exception as e:
    print(f"Có lỗi khi lưu dữ liệu: {e}")
```
